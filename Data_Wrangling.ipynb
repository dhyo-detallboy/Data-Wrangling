### Cleaning Data

1. Cleaning "customer_df"

```python
# Menghilangkan duplikat pada data "customers_df"
customers_df.drop_duplicates(inplace=True)
```

```python
# Memastikan bahwa data duplikat sudah tidak ada
print("Jumlah Duplikasi : ", customers_df.duplicated().sum())
```

```python
# Melihat kembali data yang kolom "gender" nya terdapat NaN
# Untuk menimbang metode penanganan Missing Value apa yang cocok untuk diterapkan

customers_df[customers_df.gender.isna()]
```

```python
# Menggunakan metode "Imputation" untuk menangani missing value
# Karena masih banyak data yang berguna

#menggunakan nilai yang dominan sebagai pengganti missing value tersebut
customers_df.gender.value_counts()
```

```python
# Nilai dominan pada kolom gender yaitu "Prefer not to say"
# Nilai itu yang akan digunakan sebagai pengganti missing value

customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
# Mengecek kembali apakah kolom "gender" masih terdapat "NaN"
customers_df[customers_df.gender.isna()]
```

```python
# Melihat baris data yang mengandung inaccurate value
# Yang akan menampilkan baris data dengan nilai Age maksimum

customers_df[customers_df.age == customers_df.age.max()]
```

```python
# Mengganti inaccurate value dengan nilai "70"

customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
# Mengecek ulang age max

customers_df[customers_df.age == customers_df.age.max()]
```

```python
# Masih terdapat inaccurate value, yaitu "500", maka value tsb akan diganti value dengan "50"

customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
customers_df.describe()
```

2. Cleaning "orders_df"

```python
# mengganti tipe data pada kolom order_date & delivery_date menjadi datetime

datetime_columns = ["order_date", "delivery_date"]
 
for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
# Memeriksa kembali tipe data yang sudah di revisi

orders_df.info()
```

3. Cleaning "product_df"

```python
# Menghilangkan duplikat pada data "product_df"
product_df.drop_duplicates(inplace=True)
```

```python
# Memastikan bahwa data duplikat sudah tidak ada
print("Jumlah Duplikasi : ", product_df.duplicated().sum())
```

4. Cleaning "sales_df"

```python
# melihat baris data yang mengandung missing value

sales_df[sales_df.total_price.isna()]
```

```python
# Total_price merupakan hasil perkalian antara price_per_unit dan quantity. 
# Maka cara diatas dapat digunakan untuk mengisi "total_price yang NaN"

sales_df["total_price"] = sales_df["price_per_unit"] * sales_df["quantity"]
```

```python
# Mengecek Ulang Apakah Masih Ada NaN

sales_df.isna().sum()
```
